# Visualization of the Dataset

```python
matches.hist(bins=50, figsize=(16,13))
plt.show()
```

```python
deliveries.hist(bins=50, figsize=(16,13))
plt.show()
```

```python
#getting the countplot of the season column
#this gives us the number of games played in each season
plt.figure(figsize=(12,6))
sns.countplot(x='season', data=matches)
plt.show()
```

```python
#this gives us the number of times a team has won the game
plt.figure(figsize=(12,6))
w=matches.winner.value_counts()
sns.barplot(x=w.index, y=w, data=matches, orient='v')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='season', hue='toss_decision', data=matches)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.stripplot(x='season', y='win_by_runs', data=matches, orient='v')
plt.xticks(rotation=90)
plt.show()
```

```python
#Lets see player details 
#first lets see who has scored the most runs
plt.figure(figsize=(12,6))
batsman = deliveries.groupby(['batsman'])['batsman_runs'].agg('sum').reset_index().sort_values(by='batsman_runs', ascending=False).reset_index(drop=True)
tbatsman = batsman.iloc[:10,:]
sns.barplot(x=tbatsman['batsman'], y=tbatsman['batsman_runs'], color='red')
plt.ylabel("Runs")
plt.xlabel("Players")
plt.xticks(rotation=90)
plt.show()
```

```python
#Lets see player details 
#first lets see who has scored the most sixes
plt.figure(figsize=(12,6))
batsman = deliveries.groupby(['batsman'])['batsman_runs'].agg(lambda x: (x==6).sum()).reset_index().sort_values(by='batsman_runs', ascending=False).reset_index(drop=True)
tbatsman = batsman.iloc[:10,:]
sns.barplot(x=tbatsman['batsman'], y=tbatsman['batsman_runs'], color='purple')
plt.ylabel("Runs")
plt.xlabel("Players")
plt.xticks(rotation=90)
plt.show()
```

```python
#Lets see player details 
#first lets see who has scored the most sixes
plt.figure(figsize=(12,6))
batsman = deliveries.groupby(['batsman'])['batsman_runs'].agg(lambda x: (x==4).sum()).reset_index().sort_values(by='batsman_runs', ascending=False).reset_index(drop=True)
tbatsman = batsman.iloc[:10,:]
sns.barplot(x=tbatsman['batsman'], y=tbatsman['batsman_runs'], color='violet')
plt.ylabel("Runs")
plt.xlabel("Players")
plt.xticks(rotation=90)
plt.show()
```

Now let us see the bowler data

```python
plt.figure(figsize=(12,6))
sns.countplot(x='dismissal_kind', data=deliveries)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='bowling_team', data=deliveries)
plt.xticks(rotation=90)
plt.show()
```

```python
#Lets see player details 
#first lets see which bowler with number of dot balls
plt.figure(figsize=(12,6))
batsman = deliveries.groupby(['bowler'])['total_runs'].agg(lambda x: (x==0).sum()).reset_index().sort_values(by='total_runs', ascending=False).reset_index(drop=True)
tbatsman = batsman.iloc[:10,:]
sns.barplot(x=tbatsman['bowler'], y=tbatsman['total_runs'], color='yellow')
plt.ylabel("Count")
plt.xlabel("bowler Name")
plt.title("Bowler with number of dot balls")
plt.xticks(rotation=90)
plt.show()
```

```python
#Lets see player details 
#first lets see which player has highest run scored against
plt.figure(figsize=(12,6))
batsman = deliveries.groupby(['bowler'])['extra_runs'].agg(lambda x: (x>0).sum()).reset_index().sort_values(by='extra_runs', ascending=False).reset_index(drop=True)
tbatsman = batsman.iloc[:10,:]
sns.barplot(x=tbatsman['bowler'], y=tbatsman['extra_runs'], color='green')
plt.ylabel("Runs")
plt.xlabel("bowler")
plt.xticks(rotation=90)
plt.show()
```

Now let us see the winning data

```python
#batting total wins and rate
wins = (matches.win_by_runs > 0).sum()
loss = (matches.win_by_runs ==0).sum()
plt.pie([wins,loss],labels=['wins','loss'],colors=['green','blue'], autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Wins and loss according to batting first')
plt.show()
```

```python
wins = (matches.win_by_wickets > 0).sum()
loss = (matches.win_by_wickets ==0).sum()
plt.pie([wins,loss],labels=['wins','loss'],colors=['orange','blue'], autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Wins and loss according to bowling first')
plt.show()
```
