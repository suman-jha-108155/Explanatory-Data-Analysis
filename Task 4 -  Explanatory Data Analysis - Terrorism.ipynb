# EDA

```python
#Total number of kills due to terror attack 
print(dataterr['nkill'].sum())
```

```python
#property damage in terms of US dollars
pd.set_option("display.max_rows", None)
propvalue = dataterr.groupby(['country_txt'])['propvalue'].sum().sort_values(ascending=False)
propvalue
```

```python
plt.figure(figsize=(12,6))
dataterr.groupby(['country_txt'])['propvalue'].sum().sort_values(ascending=False).head(10).plot.bar()
plt.show()
```

```python
#year with highest terror attack
plt.figure(figsize=(15,10))
dataterr['iyear'].value_counts().plot.bar(color='red')
plt.title('Year with Terror Attack')
plt.show()
```

```python
dataterr.shape
```

```python
#year and total number of kills
plt.figure(figsize=(25,10))
dataterr.groupby(['iyear'])['nkill'].sum().plot.bar(color=['black'])
plt.title('Year and Total kills')
plt.xlabel('iyear')
plt.ylabel('kills')
plt.show()
```

```python
import geopandas as gpd
# From GeoPandas, our world map data
worldmap = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))

# Creating axes and plotting world map
fig, ax = plt.subplots(figsize=(25, 35))
worldmap.boundary.plot( ax=ax)

# Plotting our kill data with a color map
x = dataterr['longitude']
y = dataterr['latitude']

plt.scatter(x, y)
# Creating axis limits and title
plt.xlim([-180, 180])
plt.ylim([-90, 90])
plt.xlabel("longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
#attack according to the regions
plt.figure(figsize=(15,10))
dataterr['region_txt'].value_counts().plot.bar(color='purple')
plt.title('Attack according to the region')
plt.show()
```

```python
#kills according to the region
plt.figure(figsize=(12,6))
sns.barplot(x="region_txt", y="nkill", data=dataterr)
plt.xticks(rotation=90)
plt.show()
```

```python
#Pie chart description of attack waepons used
plt.figure(figsize=(25,20))
dataterr['attacktype1_txt'].value_counts().plot.pie(label='Genre', autopct='%1.0f%%')
plt.show()
```

```python
#bar chart for the most targeted victims of terror attacks
plt.figure(figsize=(12,6))
dataterr['targtype1_txt'].value_counts().plot.bar(color='brown')
plt.show()
```

```python
#description of weapon used
plt.figure(figsize=(12,6))
dataterr['weaptype1_txt'].value_counts().plot.bar(color='red')
plt.title('Types of weapon used')
plt.show()
```

```python
#group which caused the most terror attacks
plt.figure(figsize=(12,6))
dataterr['gname'].value_counts().head(15).plot.bar(color='violet')
```

# Analysis of Data

The above data has been analyzed and during analysis, it is found that in 2014 there were highest number of terror attacks.
Likewise, the region with maximum terror attack are the saharan-african region and the middle east region. Similarly, Explosives
are used as the main weapon for terror attack followed by firearms. The most known group to cause the terror attacks are taliban
followed by ISIL. In the same way if see the worldmap we can see, africa asia and europe are most vulnerable to terror attacks.
We also have an insight in the target and number os victims. Private citizens and military personnel are the most targeted people
among all the victims. Likewise, United Kingdom has the most property loss follwed by USA.
